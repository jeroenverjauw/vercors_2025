```python
import os
files = [i for i in os.listdir('.') if  any(j in i for j in ['ipynb']) or float(i[4:8]) > 2030]
files
```

```python
# import subprocess
import time
import pyautogui
import pygetwindow as gw


for file in files:
    # Path to your image
    image_path = file

#     Step 1: Launch Paint with the image
    paint_proc = subprocess.Popen(["mspaint", image_path])

#     # Step 2: Wait a few seconds for Paint to open (adjust if needed)
    time.sleep(1)

    paint_window = None
    for _ in range(10):
        windows = gw.getWindowsWithTitle("Paint")
        if windows:
            paint_window = windows[0]
            paint_window.activate()
            break
        time.sleep(0.5)

#     # Step 4: Send Ctrl+S (Save)
#     time.sleep(0.5)

    # Step 4: Send Ctrl+S (Save)
    pyautogui.hotkey('ctrl', 's')
    time.sleep(0.5)
    pyautogui.hotkey('tab')
    pyautogui.typewrite('j')
    time.sleep(0.5)
    pyautogui.hotkey('enter')
    pyautogui.press('left')
    pyautogui.hotkey('enter')
    time.sleep(.5)
    print(image_path)
    # Step 5: Wait and close the window (Alt+F4)
    pyautogui.hotkey('alt', 'f4')
```
